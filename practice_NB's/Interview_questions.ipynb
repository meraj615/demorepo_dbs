```python
data=[("product1","2024-12-01",100),("product2","2024-12-01",150),("product1","2024-12-02",100),("product2","2024-12-02",200)]
columns = ["id", "date", "value"]
df=spark.createDataFrame(data,columns)
display(df)
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql.functions import col, to_date, month, sum 
```

```python
df=df.withColumn("date",to_date("date"))
df=df.withColumn("date",month("date")).groupBy("date","id").agg(sum("value"))
df=df.withColumn("rank",row_number().over(Window.partitionBy("date").orderBy(col("sum(value)").desc()))).filter(col("rank")==1)
df.display()
```
