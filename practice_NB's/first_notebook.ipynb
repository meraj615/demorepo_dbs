```python
# Create sample data for 5 years, 3 records per year
years = [2022, 2023, 2024, 2025, 2026]
staff_data = [
    (1, "Alice", "HR", 2022),
    (2, "Bob", "Finance", 2022),
    (3, "Charlie", "IT", 2022),
    (4, "David", "HR", 2023),
    (5, "Eve", "Finance", 2023),
    (6, "Frank", "IT", 2023),
    (7, "Grace", "HR", 2024),
    (8, "Heidi", "Finance", 2024),
    (9, "Ivan", "IT", 2024),
    (10, "Judy", "HR", 2025),
    (11, "Mallory", "Finance", 2025),
    (12, "Niaj", "IT", 2025),
    (13, "Olivia", "HR", 2026),
    (14, "Peggy", "Finance", 2026),
    (15, "Sybil", "IT", 2026)
]

columns = ["staff_id", "staff_name", "department", "year"]
staff_df = spark.createDataFrame(staff_data, columns)

# Create table 'staff' and insert sample records
spark.sql("""
    CREATE TABLE IF NOT EXISTS staff (
        staff_id INT,
        staff_name STRING,
        department STRING,
        year INT
    ) USING DELTA
""")
staff_df.write.format("delta").mode("overwrite").insertInto("staff")

display(spark.table("staff"))
```

```python
%sql
select * from staff
```

```python
%sql
create or replace table new_staff
using delta
partitioned by (department)
as select * from staff
```

```python
%sql
select * from new_staff
```

```python
%sql
-- Databricks SQL does not support 'clustered by' in ALTER TABLE statements.
-- Clustering is not a supported table property in Databricks SQL.
-- You can use OPTIMIZE with ZORDER for clustering-like behavior.

-- Example:
-- Issue: The column 'id' does not exist in the 'new_staff' table.
-- Use the correct column name, e.g., 'staff_id'.

OPTIMIZE new_staff
ZORDER BY (year, staff_id)
```
